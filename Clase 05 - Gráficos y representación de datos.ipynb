Una parte importante del análisis de un problema en física consiste en su visalización. Para esto usamos la librería matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Gráficos sencillos

```python
# Preparamos los datos
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

# Graficamos los datos
plt.plot(x, y)

# Mostramos el gráfico
plt.show()
```

Para producir un gráfico basta seguir 6 pasos:

* Preparar los datos
* Crear el gráfico como objeto abstracto
* Graficar
* Modificar los parámetros visuales del gráfico
* Mostrar el gráfico
* Guardar el gráfico

## Preparar los datos

Para graficar necesitamos una tabla de datos. Estos pueden ser de dos tipos:

* Resultados de la evaluación de una función (como en el ejemplo anterior).

```python
# Preparamos los datos
x = np.linspace(-1,1,100)
y1 = np.arccos(x)
y2 = np.arcsin(x)
```

```python
# Producimos el gráfico 
plt.plot(x, y1, label='$\cos^{-1}(x)$')
plt.plot(x, y2, label='$\sin^{-1}(x)$')

# Para diferenciar los dos gráficos podemos incluir una leyenda
plt.legend()

plt.show()
```

## Parámetros del gráfico

Podemos cambiar diferentes parámetros como el color, el tipo de línea o punto, etc.

```python
fig = plt.figure()
fig.add_axes()

axs1 = fig.add_subplot(2,2,1)
axs2 = fig.add_subplot(2,2,2)
axs3 = fig.add_subplot(2,2,3)

axs1.scatter(notas[:,0], notas[:,1], color='yellowgreen', marker='^')
axs2.scatter(notas[:,0], notas[:,2], color='teal', marker='o')
axs3.scatter(notas[:,1], notas[:,2], color='darkslategray', marker='+')

fig.suptitle('Correlación entre notas')
axs1.set_title('Cuántica-Clásica')
axs2.set_title('Cuántica-Experimental')
axs3.set_title('Clásica-Experimental')
fig.tight_layout(pad=2.0)

fig.show()
```

```python
# Preparamos los datos
x = np.linspace(-1,1,100)
y1 = np.arccos(x)
y2 = np.arcsin(x)
y3 = np.arctan(x)

# Producimos el gráfico 
fig = plt.figure()
fig.add_axes()
ax = fig.add_subplot(1,1,1)

ax.plot(x, y1, label='$\cos^{-1}(x)$', linestyle='--', color='turquoise', linewidth=2)
ax.plot(x, y2, label='$\sin^{-1}(x)$', linestyle='-', color='tan', linewidth=2)
ax.plot(x, y3, label='$\\tan^{-1}(x)$', linestyle=':', color='cornflowerblue', linewidth=2.5)
ax.set_xlabel('$x$')
ax.set_ylabel('Función')

# Para diferenciar los dos gráficos podemos incluir una leyenda
plt.legend()

plt.show()
```

## Describir los datos con estadísticas sencillas

```python
fig = plt.figure()
fig.add_axes()

axs1 = fig.add_subplot(2,2,1)
axs2 = fig.add_subplot(2,2,2)
axs3 = fig.add_subplot(2,2,3)

axs1.hist(notas[:,0], color='yellowgreen', bins=10)
axs2.hist(notas[:,1], color='teal', bins=10)
axs3.hist(notas[:,2], color='darkslategray', bins=10)

fig.suptitle('Histograma de notas')
axs1.set_title('Cuántica')
axs2.set_title('Clásica')
axs3.set_title('Experimental')
fig.tight_layout(pad=2.0)

fig.show()
```

Con frecuencia se describen los datos con tres estadísticas sencillas:

* Valor promedio: 
$$\mu = \frac{1}{N}\sum_i x_i$$

* Varianza ($\sigma$ se llama la desviación estándar): 
$$\sigma^2 = \frac{1}{N}\sum_i (x_i - \mu)^2$$

* Covarianza:
$$\text{Cov}(X,Y) = \frac{1}{N}\sum_i (x_i - \mu_x)(y_i - \mu_y)$$

* Coeficiente de correlación:
$$\rho(X,Y) = \frac{\text{Cov}(X,Y)}{\sigma_x \sigma_y}$$
